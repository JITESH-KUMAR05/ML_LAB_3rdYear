## visulazation using plotly and cufflinks

```python
import pandas as pd
import cufflinks as cf
import numpy as np
import plotly.express as px
```

```python
cf.go_offline()
```

```python
#Line Plot with Cufflinks
df = pd.DataFrame({'a': [4,2,3,2], 'b': [4,5,6,5]})
df.iplot(kind='line', title='Line Plot with Cufflinks')
```

```python
df.head()
```

```python
df.iplot(kind='line',title='line plot using plotly and cufflinks')
```

```python
fig = px.line(df, title='line plot with the plotly express')
fig.show()
```
