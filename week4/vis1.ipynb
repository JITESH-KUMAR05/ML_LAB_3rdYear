## parametric and non parametric

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
x = np.linspace(0,10,100)
y = np.sin(x) + np.random.normal(scale=0.3,size=100)
```

```python
lowers_res=sm.nonparametric.lowess(y,x,frac=0.2)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y,label='Data',alpha=0.6)
```

```python
plt.plot(lowers_res[:,0],lowers_res[:,1],color='red',label='Lowess')
```

```python
plt.title('locally weighted regression(Lowess)')
```

```python
plt.legend()
```

```python
plt.show()
```

```python
plt.scatter(x,y,label='Data',alpha=0.6)
plt.plot(lowers_res[:,0],lowers_res[:,1],color='red',label='Lowess')
plt.title('locally weighted regression(Lowess)')
plt.legend()
plt.show()
```
